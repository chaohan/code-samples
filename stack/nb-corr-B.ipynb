```python
from stack import *

N = 200; 
L = 100.0e3; dL=0.0*L; 
L1= 300.0e3;  dL1=0.000*L1; 
n0= 1.0; n1=1.35

stacks = map(lambda q:random_stack(N,n0,n1,dL,L,dL1=dL1,L1=L1),[1])#[1.0,0.5,0.25,0.1,0.01,0.001]
pyplot.figure(figsize=(18,5))
stacks[0].plot()
```

```python
#wlrange = [700.0,900.0]
wlrange = [1595.0,1615.0]
Nmodes = 1000
dw = (wlrange[1]-wlrange[0])/float(Nmodes)
#print 1.0/dw,"data points per nm"
#print dw,"nm per data point"
```

```python
specs = map(lambda s:s.scan('transmission',wlrange,Nmodes,style='-'),stacks)
```
