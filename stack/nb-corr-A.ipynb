<p> What I've done here is to show how the thickness of air layers changes the photonic bands of the stack. And I believe that, ultimately, the photonic band structure is what controls our focusing ability.</p>

<p> The following are 6 stack samples with <i>no thickness variation</i> in either type of layers. This is to show the bands more clearly. When there's significant variation, periodic structures are expected to follow these bands.</p>

<p> The air gap thicknesses for the following 6 stacks are respectively: 100, 50, 25, 10, 1, 0.1 micron, with the last one close to what you have (0.14 micron). All other parameters are the same as yours. </p>
<br/>
</font>

```python
from stack import *

N = 60; 
L = 100.0e3; dL=0.0*L; 
L1= 100.0e3;  dL1=0.000*L1; 
n0= 1.0; n1=1.5

stacks = map(lambda q:random_stack(N,n0,n1,dL,L,dL1=dL1,L1=q*L1),[1.0,0.5,0.25,0.1,0.01,0.001])

def costum_plot(s):
    pyplot.figure(figsize=(18,5))
    s.plot()

output = map(costum_plot,stacks)
```

<font size="4.5">
<p> Now I'm scanning the transmission spectram around 800nm like you did). </p>

<p> As you can see below (where air gap = 100, 50, 25 microns), the <b>periodicity wavelength</b> is really where we have <b>exactly repeated structure</b> (in the first case, it is the set of one big bump followed by three thinner ones). So what's most important is that each <b>fully periodic structure</b> has several <b>sub-passband</b>. </p>

<p> And when we decrease the air gap thickness, such periodicity length of courses increases. </p>
<br/>

</font>

```python
wlrange = [750.0,850.0]
#wlrange = [400.0,405.0]
Nmodes = 2000
dw = (wlrange[1]-wlrange[0])/float(Nmodes)
#print 1.0/dw,"data points per nm"
#print dw,"nm per data point"
```

```python
specs = map(lambda s:s.scan('transmission',[800.0,830.0],Nmodes,style='-'),stacks[0:3])
```

<br/>
<br/>
<font size="4.5">

<p> Now when the air gaps are even smaller (10, 1, 0.1 microns), the periodicity length gets really large (not even shown in the range of the scan in the 1-micron and 0.1-micron cases). </p>

<p> Meanwhile, what's more important, is that <b>the number of sub-passbands is increasing</b>. And eventually, <b>they become really periodic themselves</b>. In the case when air gap is below 1 micron, we can clearly see the sub-passbands getting very periodic, at around 2nm periodic length, like you discovered. </p>

<p> I believe this explains my problem exactly (plots are at the bottom): With my choice of 300 micron air gaps, I estimate my periodicity length is over 5nm. But what I missed is that there are in fact a couple of sub-passbands within this fully periodic range! That's why I lose focusing ability when I get close to 1nm resolution (which is more than half the smallest sub-passband).</p>
</font>

<font size="5">
<p> So I belive the conclusion is really this: we have control/focus ability as long as our shaping resolution is well within the width of the smallest passbands (as opposed to the exact periodicity length). </p>
</font>
